#### (i) Define funciones para hacer un solo paso de los métodos de Euler, punto medio, y RK4. Aceptan la función que integrar, el valor actual de la variable independiente, y el tamaño de paso, y regresan el nuevo valor de la variable independiente.

```julia
function Euler(f,t, h, x =0)
    h*f(t, x)
end
```

```julia
function punto_medio(f,t,h, x = 0)
    ((h/2)*(f(t,x)+ f(t+h,x+(h*f(t, x)))))
end 
```

```julia
function RK4(f, t, h,x=0)
    k1 = h*f(t, x)
    k2 = h*f(t + h/2, x + k1/2)
    k3 = h*f(t + h/2, x + k2/2)
    k4 = h*f(t + h, x + k3)
    (k1 + 2*k2 + 2*k3 + k4)/6
end
```

#### (ii) Define una función integrar que acepta un método de integración que usar, una función, el rango de tiempos y el tamaño del paso

```julia
function integrar(metodo,f,t0,h)
    T=0
    a=Any[]
    resultado=0
    for t = (t0[1]):h:t0[2]
        T = metodo(f,t,h)
        resultado+=T
        push!(a,resultado)
    end
    a, resultado
end
```

```julia
function f(x, t)
    cos(x)
end
```

#### (iii) Compara el funcionamiento de estos 3 métodos para el oscilador armónico.

Definimos ahora Euler,punto_medio y RK4 de forma que nos permitan solucionar sistemas de escuaciones

```julia
function Euler2(f, t, x, v, h)
    k1 = h*f(t, x, v)
    l1 = h*v
    [k1 , l1]
end
```

```julia
function punto_medio2(f, t, x, v, h)
    k1 = h*f(t, x, v)
    l1 = h*v
    k2 = h*f(t + h/2 , x + l1/2 , v + k1/2)
    l2 = h*(v + k2/2)
    [(k1 + k2)/2 , (l1 + l2)/2]
end
```

```julia
function RK42(f, t, x, v, h)
    k1 = h*f(t, x, v)
    l1 = h*v
    k2 = h*f(t + h/2, x + l1/2, v + k1/2)
    l2 = h*(v + k2/2)
    k3 = h*f(t + h/2, x + l2/2, v + k2/2)
    l3 = h*(v + k3/2)
    k4 = h*f(t + h, x + k3, v + l3)
    l4 = h*(v + k4)
    [(k1 + 2*k2 + 2*k3 + k4)/6, (l1 + 2*l2 + 2*l3 + l4)/6]
end
```

```julia
function Oscilador(metodo, x0, v0, t0, tf, h, k )
    X = Float64[x0]
    V = Float64[v0]
    T = Float64[t0]
    for t = t0+h:h:tf
        function f(t, x, v)
            -k*x
        end
        v_nueva = v0 + metodo(f, t, x0, v0, h)[1]
        x_nueva = x0 + metodo(f, t, x0, v0, h)[2]
        x0 = x_nueva
        v0 = v_nueva
        push!(V, v0)
        push!(X, x0)
        push!(T, t)
    end
    T, X, V
end
```

Resolvemos el oscilador para los tres métodos

```julia
t_e, x_e, v_e = Oscilador(Euler2, 5, 0, 0, 100, .01,1);
t_pm, x_pm, v_pm = Oscilador(punto_medio2, 5, 0, 0, 100, .01,1);
t_rk4, x_rk4, v_rk4 = Oscilador(RK42, 5, 0, 0, 100, .01,1);
```

```julia
plot(t_rk4,x_rk4,label="RK4")
plot(t_e,x_e,label="Euler")
plot(t_pm,x_pm,label="Punto Medio")
legend()
```

#### (v) Resuelva numéricamente el problema de los dos carritos de la tarea 5.

```julia
function resolver(metodo,f, x0, v0, t0, tf, h )
    X = Vector{Float64}[[x0]]
    V = Vector{Float64}[[v0]]
    T = Float64[t0]
    for t = t0+h:h:tf
        v_nueva = v0 + metodo(f, t, x0, v0, h)[1:length(x0)]
        x_nueva = x0 + metodo(f, t, x0, v0, h)[length(x0)+1:2*length(x0)]
        x0 = x_nueva
        v0 = v_nueva
        push!(V, v0)
        push!(X, x0)
        push!(T, t)
    end
    T, X, V
end
```

```julia
function carrito1(t, x, v)
    k=1
    [(-k*x[1]) + k*(x[2]-x[1]), (-k*x[2]) - k*(x[2]-x[1])]
end
```

```julia
t_c1, x_c1, v_c1 = resolver(RK42, carrito1,[-3,3], [0,0], 0, 10, .01);
```

```julia
plot(t_c1,x_c1)
```

```julia
t_c2, x_c2, v_c2 = resolver(RK42, carrito1,[0,3], [0,0], 0, 10, .01);
plot(t_c2,x_c2)
```
